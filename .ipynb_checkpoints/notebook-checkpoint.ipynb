```python
def chunkify(filepath, chunk_size, outputname, timing=False):
    batch_no = 1
    for chunk in pd.read_json(filepath, chunksize=chunk_size, lines=True, compression='bz2'):
        if timing:
            before = time.time()

        output = 'Data/' + outputname + '-' + str(batch_no) + '.csv'

        chunk.to_csv(output, index=False)

        compression_level = 9
        source_file = output
        destination_file = output + '.bz2'

        with open(source_file, 'rb') as data:
            compressed = bz2.compress(data.read(), compression_level)
        fh = open(destination_file, "wb")
        fh.write(compressed)
        fh.close()

        os.remove(output)

        if timing:
            after = time.time()
            print(after - before, 's')

        batch_no += 1
```

```python
def chunkify(filepath, chunk_size, outputname, timing=False):
    batch_no = 1
    chunks = pd.read_json(filepath, chunksize=chunk_size, lines=True, compression='bz2')
    if timing:
        before = time.time()
    
    def innerChunk(chunk):
        output = 'Data/' + outputname + '-' + str(batch_no) + '.csv'

        chunk.to_csv(output, index=False)

        compression_level = 9
        source_file = output
        destination_file = output + '.bz2'

        with open(source_file, 'rb') as data:
            compressed = bz2.compress(data.read(), compression_level)
        fh = open(destination_file, "wb")
        fh.write(compressed)
        fh.close()

        os.remove(output)

        if timing:
            after = time.time()
            print(after - before, 's')

        batch_no += 1
    
    with ThreadPoolExecutor() as exe:
        exe.map(innerChunk, chunks)
```

```python
chunkify('Data/quotes-2016.bz2', 50000, '2016', timing=True)
```
